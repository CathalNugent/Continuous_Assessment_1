# Section 1: Data Preparation

```python
#Install relevant liabraries to be used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Population.csv")
```

```python
df.head()
```

```python
#Now let just remove any columns that we don't need
columns_to_remove = ['Country of origin (ISO)', 'Country of asylum (ISO)', 
                     'IDPs of concern to UNHCR', 'Other people in need of international protection',
                     'Stateless persons', 'Host Community', 'Others of concern']


df1 = df.drop(columns = columns_to_remove)
```

```python
#Ranming the columns 
new_column_names = {"Refugees under UNHCR's mandate": 'Refugees'}
df1.rename(columns=new_column_names, inplace=True)

df1
```

```python
# Adding in a total column to sum both Refugees and Asylum-seekers
df1['Total'] = df1['Refugees'] + df1['Asylum-seekers']
df1
```

# Section 2: Grouping by Year

```python
# Grouping by Year and summing up the values
Year_df = df1.groupby('Year').sum()

Year_df
```

```python
#Plotting our Data
Year_df.plot(y=['Refugees', 'Asylum-seekers'], kind='line', marker='o', linestyle='-', figsize=(12, 6))

# Adding labels and title
plt.title('Number of Refugees and Asylum-seekers by Year')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.ylim(bottom=0)

plt.show()
```

I think the latest year skews the view of the graph slightly. So I have decided to replot this graph with 2022 year removed so we can get a closer look at the prior years. 

```python
Year_df_filtered = Year_df[Year_df.index != 2022]

# Plotting the filtered data
Year_df_filtered.plot(y=['Refugees', 'Asylum-seekers'], kind='line', marker='o', linestyle='-', figsize=(12, 6))

# Adding labels and title
plt.title('Number of Refugees and Asylum-seekers by Year (excluding 2021)')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.ylim(bottom=0)

plt.show()
```

# Section 3: Grouping by Country of Origin

Section 3.1: Refugees and Asylum-seekers in 2022

```python
#Select a specific year as the data is in stock format
df_2022 = df1[df1['Year'] == 2022]
#df_2022
```

```python
small_countries = df_2022['Country of origin'][df_2022['Total'] < 500].tolist()
df_2022.loc[df_2022['Country of origin'].isin(small_countries), 'Country of origin'] = 'All Other Countries'
df_2022 = df_2022.groupby('Country of origin', as_index=False).sum()
```

```python
# Define a custom color scheme
custom_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', 
                 '#ffb3e6', '#c2f0c2', '#ffb366', '#c2c2d6']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_2022['Total'], labels = df_2022['Country of origin'], 
        autopct=lambda p: '{:.1f}%'.format(p) if p > 1 else '', 
        startangle=140, colors=custom_colors)

plt.title('Distribution of Refugees by Country of Origin')
plt.show()
```

Section 3.2: Refugees and Asylum-seekers in 2021

```python
#Select a specific year as the data is in stock format
df_2021 = df1[df1['Year'] == 2021]
#df_2021
```

```python
small_countries = df_2021['Country of origin'][df_2021['Total'] < 300].tolist()
df_2021.loc[df_2021['Country of origin'].isin(small_countries), 'Country of origin'] = 'All Other Countries'
df_2021 = df_2021.groupby('Country of origin', as_index= False ).sum()
```

```python


# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_2021['Total'], labels = df_2021['Country of origin'], 
        autopct=lambda p: '{:.1f}%'.format(p) if p > 1 else '', 
        startangle=140, colors=custom_colors)

plt.title('Distribution of Refugees by Country of Origin')
plt.show()
```

Section 3.2: Refugees and Asylum-seekers from Palestine & Israel

```python
#Select a specific year as the data is in stock format
Pal_df = df1[df1['Country of origin'] == 'Palestinian']
Isr_df = df1[df1['Country of origin'] == 'Israel']
#Pal_df
```

```python
# Plotting
plt.figure(figsize=(14, 6))
plt.plot(Pal_df['Year'], Pal_df['Total'], label='Palestinian', marker='o')
plt.plot(Isr_df['Year'], Isr_df['Total'], label='Israeli', marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.title('Refugees and Asylum-seekers from Palestine/Israel (1999-2022)')
plt.legend()
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.ylim(bottom=0)

# Display the plot
plt.show()
```
